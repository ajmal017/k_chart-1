```python
import numpy as np
import pandas as pd
from pandas_datareader import data as web
from datetime import datetime as dt
```

```python
from KChart import k_chart
```

```python
# Read and Load Data
#start = dt(2017,1,1)
#end = dt(2017,11,30)
#ticker = 'SPY'
#source_data = web.DataReader(ticker, 'yahoo', start, end)
source_data = pd.read_csv('spy.csv', parse_dates=True, index_col=0)
```

```python
# Calc
df=source_data.copy()
# calc sub_indicator
df['diff'] = df['Adj Close']/df['Close']
# calc main_indicator
df['ma'] = df['Close'].rolling(12,center=True,min_periods=1).mean()
```

```python
# Plot Data
```

```python
# plot ohlc candlestick
df.k_chart()
```

```python
# add volume
df.k_chart(volume_col='Volume')
```

```python
# add sub_indicator
df.k_chart(sub_indicator_cols='diff')
```

```python
# add main indicator
df.k_chart(main_indicator_cols='ma', volume_col='Volume', sub_indicator_cols='diff')
```
