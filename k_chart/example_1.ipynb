```python
# Read and Load Data
```

```python
start = dt(2017,1,1)
end = dt(2017,11,30)
ticker = 'SPY'
source_data = web.DataReader(ticker, 'yahoo', start, end)
```

```python
df=source_data.copy()
```

```python
df['diff'] = df['Adj Close']/df['Close']
```

```python
df['ma'] = df['Close'].rolling(12,center=True,min_periods=1).mean()
```
